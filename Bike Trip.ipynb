```python
#Dependencies
import pandas as pd
```

```python
#Pull in CSV data for Jan1-Jun30 2020
startof2020 = pd.concat(map(pd.read_csv, ['./CSVs/202001-citibike-tripdata.csv',
                                     './CSVs/202002-citibike-tripdata.csv',
                                     './CSVs/202003-citibike-tripdata.csv',
                                     './CSVs/202004-citibike-tripdata.csv',
                                     './CSVs/202005-citibike-tripdata.csv',
                                     './CSVs/202006-citibike-tripdata.csv']))
```

```python
#Preview DataFrame
startof2020
```

```python
#Pull the columns needed
tripID = startof2020[["tripduration","bikeid"]]
tripID
```

```python
#Check column types
tripID.dtypes
```

```python
#Groupby Bike ID and add the trip duration column
bikeID = tripID.groupby(["bikeid"]).sum()
```

```python
#Preview DataFrame
bikeID
```

```python
#Add a column breaking the time into minutes(Originally seconds)
bikeID["triptime(min)"] = bikeID["tripduration"]/60
```

```python
#The calculated pace from citibike is 7.5 MPH which is an 8-minute mile so find the miles using that
#      https://twitter.com/CitiBikeNYC/status/920315023938981888
bikeID["miles"] = bikeID["triptime(min)"]/8
```

```python
#Preview DataFrame
bikeID
```

```python
#Find the Average Miles for Reference later in Tableau
bikeID["miles"].mean()
```

```python
#Save to a CSV to inport into Tableau
bikeID.to_csv('bikeMiles.csv', index=False)
```
